### Reading in semi-structured text data

- source: https://archive.ics.uci.edu/ml/machine-learning-databases/00228/ 
- other databases: https://archive.ics.uci.edu/ml/machine-learning-databases/

```python
import pandas as pd

data = pd.read_csv('dataset/SMSSpamCollection.csv', sep='delimiter')
data.head()
```

```python
data.shape
data.columns
```

```python
# https://stackoverflow.com/questions/37333299/splitting-a-pandas-dataframe-column-by-delimiter
messages = pd.DataFrame(columns=['label','text'])
messages[['label','text']] = data['v1\tv2'].str.split('\t', expand=True)
```

### removing punctuation

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    text = ''.join([char for char in text if char not in string.punctuation])
    return text


### applying the function with lambda
messages['text_clean'] = messages['text'].apply(lambda x: remove_punctuation(x))

messages.head()
```

---------------------------------------------------

### tokenization

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

messages['text_tokenized'] = messages['text_clean'].apply(lambda x: tokenize(x.lower()))
messages.head()
```

------------------------

### stopwords

```python
import nltk
stopwords = nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_text):
    text = [word for word in tokenized_text if word not in stopwords]
    return text

messages['text_nonstop'] = messages['text_tokenized'].apply(lambda x: remove_stopwords(x))

messages.head()
```
